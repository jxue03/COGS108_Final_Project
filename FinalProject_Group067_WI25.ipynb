# Abstract

In the United States, jails have increasingly become de facto mental health institutions, disproportionately housing individuals with untreated mental health conditions. Motivated by the Data-Driven Justice Initiative and the urgent need to address mental health crises outside of the criminal justice system, this project explores the relationship between mental health treatment accessibility and jail admission rates across California counties from 2020 to 2024.

We integrated data on the number of mental health facilities per county with incarceration data to evaluate potential correlations. Using linear regression, we assessed the impact of treatment availability on jail admissions while accounting for confounding socioeconomic factors.

Our findings suggest a negative correlation between access to mental health care and jail admission rates: counties with more mental health facilities generally exhibited lower incarceration rates. However, this relationship is nuanced and heavily influenced by variables such as median income and homelessness rates. While more research should be done to understand the underlying factors of the relationship we found, this study supports the idea that expanded mental health infrastructure is a potential avenue in broader criminal justice reform efforts.

# Research Question

Does greater access to mental health care correlate with lower jail admission rates? Specifically, how does the amount of mental health treatment facilities in a county impact county-level incarceration rates in California from 2020 to 2024?

## Background and Prior Work

Our research question was inspired by a talk given by DJ Patil, the former U.S. Chief Data Scientist, in which he discussed the Data-Driven Justice Initiative (DDJ). The DDJ is an initiative that worked to reduce recidivism – the tendency of someone previously incarcerated to reoffend and end up back in jail again – particularly for those struggling with mental health crises. Many individuals cycle in and out of jail for low-level offenses driven by mental health issues and substance abuse. Data from the Obama administration’s initiative shows that approximately **64%** of incarcerated individuals experience mental illness, while **68%** struggle with substance abuse.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Significant taxpayer money is being spent on cycling individuals in and out of jail, but more importantly, these people are stuck in a never-ending loop of incarceration that could be addressed with mental health services. This system worsens their mental health, narrowing their opportunities and stripping away chances for stability and reintegration. Instead of support, they face punishment, leaving them with fewer options and a higher chance of returning to jail.

This practice of housing the mentally ill in incarceration facilities is a relatively recent phenomenon. In the 1950s, the deinstitutionalization movement led to public mental health institutions being shut down to transfer the mentally ill to be housed and taken care of elsewhere, originally thought to be their families. However, this transfer did not occur as fully intended; instead, with the criminalization of mental illness that occurred in tandem with deinstitutionalization, former patients of these institutions were funneled into the prison system, effectively turning jails into the nation’s new mental healthcare facilities.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Since its implementation in 2016 under the Obama administration, the DDJ has demonstrated promising results. For instance, in Miami, law enforcement officers were trained in mental health de-escalation techniques, leading to significant outcomes over five years:<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)
* Over 10,000 individuals diverted from jail to treatment
* Only 109 arrests out of 50,000 mental health-related calls
* A 40% reduction in jail population, enabling the closure of an entire county jail facility
* A drop in recidivism rates from 72% to 20% among those receiving mental health treatment

Although the initiative is no longer explicitly active on the federal level, the DDJ helped coordinate efforts at the state and local levels that ensure their principles continue to live on today.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Given these findings, our project seeks to determine whether access to mental health resources influences incarceration rates across different regions. It is important to note that while we recognize the highly intertwined nature of mental health and substance abuse issues, our project will focus solely on mental health, and as such we will not be addressing substance abuse directly.

Overall, our project aims to uncover patterns in the relationship between mental health resource availability and incarceration rates. Ultimately, our findings may offer valuable insights into policy improvements and systemic reforms that could reduce recidivism and better support individuals with mental health challenges.

## Data Wrangling

The end goal of our wrangling is to combine all of our variables - jail, mental health facilities, income, and unemploymet - into one big dataset that we can then perfom analysis on. In order to get there, we first have to wrangle each of our datasets individually.

### Population Data

As we saw earlier when importing the dataset, our population dataframe has some unwanted rows and columns, so we will start by eliminating those and renaming our wanted columns to be informative.

```python
# remove non-data rows
population = population.drop([0,1,2,3])
population = population.dropna()
population = population.drop('Unnamed: 1',axis=1)

# rename columns
population.columns=['COUNTY', '2020', '2021', '2022', '2023']
population.reset_index(inplace=True, drop=True)

population.head()
```

Now, we need to clean up the county names. This column will be the one we use to merge all our datasets in the end, so we want the formatting to be consistent across all of them. The format we’re aiming for is just the county name, without "County" or anything else added.

```python
# establishing function to standardize county name in all datasets
def standardize_county(string):
    """
    Standardizes county name strings based on the specific formats found in our datasets.

    This function is designed to handle cases such as:
    - Leading symbols (e.g., '.')
    - The presence of 'County' or ', California'
    - Trailing data like '!!Percent'
    - Fully capitalized county names

    It cleans the input string and returns the standardized county name.
    """
    
    # eliminate leading & trailing info
    string = string.lstrip('.').strip()
    string = string.replace('County', '')
    string = string.replace(', California', '')
    string = string.split('!!')[0]  # removes '!!' and anything after
    string = string.strip()

    # if all caps, convert to title case
    if string.isupper():
        string = string.title()

    return string
```

```python
# reformat counties
population['COUNTY'] = population['COUNTY'].apply(standardize_county)

population.head()
```

```python
population.dtypes
```

Since we see commas in the numeric data in our year columns, we’ve checked and found that all the year data is actually in string format, but we need it to be numeric. So, we’ll convert it to a numeric format.

```python
# remove commas and convert to int for year columns
cols = ['2020', '2021', '2022', '2023']
population[cols] = population[cols].replace(',', '', regex=True).astype(int)

population.head()
```

```python
population.dtypes
```

As we can see, we have successfully made our year data into integers rather than strings. Now, we want to average the population data across all of the years and condense it into one column.

```python
# averaging population across years
population["avg_population"] = population.iloc[:, 1:5].mean(axis=1)

# putting into concise dataset for merging
population_clean = population.drop(columns=["2020", "2021", "2022", "2023"])
population_clean = population_clean.rename(columns={"COUNTY": "county_name"})

population_clean.head()
```

```python
population_clean.shape
```

We end up with a clean dataset with population data for all 58 counties in California, ready for merging. 

### Jail Data

Since the original dataset has information for all of the U.S., we will start by filtering the jail data to contain only information about **California** jails.

```python
# filtering for only CA
ca_jail_data = jail_population[jail_population['state_name'] == 'California']

ca_jail_data.head()
```

Now, we will isolate the **county** and **year** from the California jail population data. Note that while we will not be including the **urbanicity_type** variable in this analysis, it could be a useful variable to look into in an extension of this project.

```python
# getting year from 'date_name' column
ca_jail_data = ca_jail_data.assign(year=ca_jail_data['date_name'].str[:4].astype(int)) # extracts the year from first 4 characters

# group by year and county, then calculate the average jail population
avg_jail_pop = ca_jail_data.groupby(['year', 'county_name'])['jail_population'].mean().reset_index()

# rename column for clarity
avg_jail_pop.rename(columns={'jail_population': 'avg_jail_population'}, inplace=True)

avg_jail_pop.head()
```

Now, we want an **average jail population** variable that averages across all the years. 

```python
# calculate average across years
avg_jail_pop = avg_jail_pop.groupby("county_name", as_index=False)["avg_jail_population"].mean()

avg_jail_pop.head()
```

We also see that our county names still have "County" trailing them, so we will get them into the proper format for our later merging.

```python
# reformat counties
avg_jail_pop["county_name"] = avg_jail_pop["county_name"].apply(standardize_county)

avg_jail_pop.head()
```

```python
avg_jail_pop.shape
```

Looking at the shape of our new datset, it appears that the original was missing jail information from 2 counties in California: Alpine and Sierra. This is because Alpine and Sierra counties in California lack traditional county jail facilites, and instead those incarcerated are housed in neighboring county's jails.

Now we want to see the *incarceration rate*. To do this, we will divide the **avg_jail_population** column by the **avg_population** inside the **population_clean** dataset. The result will be a proportion of the population rather than raw numbers, giving us the percentage of each county's population that is incarcerated in jail. This way we are controlling for population in our analysis.

```python
# merging population data by county
incarceration_rate = pd.merge(avg_jail_pop, population_clean, on="county_name", how="outer")

# calculating incarceration rate (%)
incarceration_rate["incarceration_rate"] = incarceration_rate["avg_jail_population"] / incarceration_rate["avg_population"] * 100

# isolating incarceration rate
incarceration_rate = incarceration_rate.drop(columns=["avg_jail_population", "avg_population"])

incarceration_rate.head()
```

Finally, we have our fully prepared **incarceration rate** dataset, shown above. Notably, our incarceration rates seem quite small, however it makes sense when cotextualized; for example, Alameda has an incarceration rate of about 0.1%, meaning roughly 1 in every 1,000 people in Alameda is incarcerated.

### Unemployment Data

To start, our data is already isolated to California, however there are  alot of unnecessary rows and columns we don't need. Since we want to get the unemployement *rate* for each California county, we will first isolate the **unemployed** row along with only columns for the **percent** estimate for each county.

```python
# keeping only unemployed row
unemployment = unemployment.iloc[[5]]

# keeping only county % columns
unemployment = unemployment.filter(regex=r"County, California!!Percent(?! Margin of Error)") # also eliminating margin of error columns

unemployment.head()
```

With our columns of interest isolated, we now want to make our dataset have county names as observations rather than variables and unemployment rate as a variable rather than an observation. 

```python
# flip rows & columns, reset index
unemployment_rate = unemployment.transpose().reset_index()

# renaming columns
unemployment_rate.columns = ["county_name", "unemployment_rate"]

unemployment_rate.head()
```

```python
unemployment_rate.shape
```

One thing to note is that despite the fact that this is census data, it only contains information for 42 out of the 58 total counties in California. While 16 missing counties is a lot, we included unemployment rate as a confounding variable, so it's not a major concern.

```python
# reformat counties
unemployment_rate["county_name"] = unemployment_rate["county_name"].apply(standardize_county)

# remove % from unemployment rate and make numeric
unemployment_rate["unemployment_rate"] = unemployment_rate["unemployment_rate"].str.rstrip('%').astype(float)

unemployment_rate.head()
```

Finally, we have our fully prepared unemployment rate dataframe, ready for merging.

## OLS Regression Analysis

In order to really understand the relationship between incarceration rate and total number of mental health facilities, we will use an OLS regression. 

### Simple Model: Incarceration Rate & Mental Health Facilities

First, we will start by examining the relationship between solely incarceration rate and total number of mental health facilities. 

```python
formula = "incarceration_rate ~ total_facilities"
model = smf.ols(formula=formula, data=df_merged).fit()
print(model.summary())
```

Examining the results of the model, from the R-squared value, we can see that the number of mental health facilities in a county accounts for about 14% of the variability in incarceration rate. So, mental health facility number does play a role in incarceration rate, however there are other factors at play as well. 

We can also see that there is a negative relationship between incarceration rate and number of mental health facilties in a county, and this relationship is statisically significant with a p-value of **0.004**.

Breaking it down further, we can see that the coefficient value for total_mental_facilties shows a relationship of -0.0147, meaning that for every additional mental health facility in a county, the incarceration rate decreases by 0.0147. So, the fact that the coefficient is negative suggests that having more mental facilities does decrease incarceration rate. As for the magnitude of the coefficient, while the coefficient seems incredibly small and thus seems like having more mental health facilties doesn't decrease incarceration by much, when viewed in the context of our incarceration rates the effect is much more than it seems. 

```python
print(df_merged['incarceration_rate'].min())
print(df_merged['incarceration_rate'].max())
```

So as you can see, in the numerical context of our data, a coefficient of -0.0147 would mean that for the county with the lowest incarceration rate, according to our model results, adding one mental health facility would decrease that county's incarceration rate by **19.9%**. On the other hand, for our highest incarceration rate county, adding one mental health facility would decrease that county's incarceration rate by **4.4%**. So, while the impact of adding mental health facilities is greatest in counties with lower incarceration rates compared to high, the impact is significant nonetheless. 

### Full Model: Incarceration Rate, Mental Health Facilities, Income, & Unemployment

We also want to analyze our results in the context of our confounding variables that we included in our merged dataset, income and unemployment. So, for our final analysis, we will use an OLS model on our full merged dataset.

However, because of the high correlation between incarceration rate and median household income, which we saw previously was a strong -0.71, we cannot include the raw median household income data due to **multicollinearity**. Multicollinearity distorts coefficient estimates, making it hard to tell which factor is actually driving the results because the variables are too closely related. This would reduce our model reliability, and so in order to still include the variable in our model, we will *log transform* median household income. By log transforming, we reduce multicollinearity by compressing the scale of highly skewed data, shrinking large differences between values. This makes the relationship between income and other variables more proportional to each other in our regression.

```python
# log transforming income to reduce multicolinearity
df_merged["log_income"] = np.log(df_merged["median_income_family"])
```

Now, we can run the full model using the log transformed income variable rather than the original income variable.

```python
# doing regression on full merged data
formula = "incarceration_rate ~ total_facilities + log_income + unemployment_rate"
model = smf.ols(formula=formula, data=df_merged).fit()
print(model.summary())
```

Looking at the results, we have an R-squared of 0.542, meaning that the variables in our model (number of mental health facilities, income, and unemployment rate) account for about **54%** of the variability in incarceration rate. This is significantly more than in our previous simple model, meaning that our other variables added a lot of explanation for incarceration rates relative to only number of mental health facilities.

Importantly, total mental facilities is *no longer significant*, with a p-value of **0.406**. The coefficient is also a lot smaller than before, with a value of -0.0032, indicating a lower amount of power that number of mental health facilities has in determining incarceration rate in a county. 

We can also see that even when log transformed, income has the most weight (largest effect) in determining incarceration rate for a county in our model. The p-value is so low it is registered as 0.000 in our model results, and there is a heavy negative coefficient of -0.1372. This tells us that counties with higher median household income tend to have lower incarceration rates, supporting our previous idea that economic stability plays a major role in reducing incarceration. 

Also of note is unemployment rate, which has a coefficient of 0.0108 and a p-value of 0.221. The positive coefficient indicates that higher unemployment rates are linked to higher incarceration; however, this relationship is not statistically significant. The magnitude of the coefficient tells us that comparatively, unemployment rates explain incarceration rates more than total number of mental health facilities but less than median household income. It is also possible that income having such a heavy effect on incarceration rate may also be weighing down the impact of unemployment rate.

### Comparing the Simple vs. Full Models

Comparing the two models, we can see that without other confounding variables, the number of mental facilities in a county are significant in predicting incarceration rate. However, in the full model, it becomes insignificant. This means that income and unemployment – income in particular – explains more of the variance for incarceration rate than any other variable. So mental health facilities seem to matter when analyzed alone, but once we control for economic factors, the impact fades, as economic factors influence incarceration rates more than mental health facility amount alone.

In terms of our research question, we already know that incarceration rate has many factors influencing it; our goal is not to see what factors underlie incarceration rate in a county - we know that many factors do, and that the number of mental health facilities in said county is likely not the primary one. What we did want to find was whether or not and to what extent number of mental health facilities influence incarceration rate. What we ended up finding was that, while our simple model suggested a significant relationship, the full model showed that this effect pales in comparison to the other main factors that drive incarceration rates. 

What we really wanted to get at with this project is that since a large amount of people with mental illness are being housed in jail rather than appropriate mental health institutions since the deinstitutionalization movement in the 50s, places with more mental health institutions would theoretically utilize them and funnel people appropriately into these facilties rather than jail. 

Ultimately, we found that we cannot say for sure that places with more mental health facilities are actually housing those with mental health issues in them over housing them in jail; there are other barriers such as money that affect access to actually utilizing these facilities that appear to moderate this relationship significantly. In future projects that extend this question, access can be defined differently to find more influential results.

# Ethics & Privacy

To protect individual privacy, all data used in this study is anonymized and generalized, ensuring that no personal information can be traced back to specific individuals. Ethically, we recognize that the data could unintentionally suggest that certain communities perform worse even when mental health care is accessible. This may not accurately reflect the true impact of mental health care if those communities face broader systemic issues, such as lower standards of living. Additionally, the data might imply that access to mental health care is ineffective if incarceration rates remain high, without accounting for factors like crime severity or structural inequalities. Bias is another key concern; wealthier areas are often better equipped to provide mental health treatment and may naturally have lower incarceration rates, skewing the results. Furthermore, mental health care may not address the root cause of every crime, especially in cases related to non-violent offenses like drug possession. Our analysis does not include data on substance abuse or substance-related offenses, which could significantly impact both mental health outcomes and jail incarceration rates. Since many of our datasets were gathered from the U.S. Census, there is also likely a lack of proper representation for homeless populations in our data, groups that are often disproportionately affected by both mental health issues and incarceration. To mitigate these biases, we incorporated factors such as local income levels and unemployment rates to provide a more balanced and accurate interpretation of the findings. However, we acknowledge that some implicit biases or gaps in representation may remain in the datasets we used that we were not able to fully overcome.

# Conclusion

Our research sought out to investigate whether greater access to mental health treatment facilities correlates with lower incarceration rates in California counties from 2020 to 2024. Inspired by DJ Patil’s lecture on the Data-Driven Justice Initiative (DDJ), we sought to examine the viability of providing mental health services in order to reduce recidivism. This serves as an alternative to the endless taxpayer money being spent on the prison industry where many individuals are stuck in the never-ending loop of incarceration. We hypothesized that counties with a greater number of mental health treatment facilities will have lower incarceration rates. 

Our findings reveal a moderate negative correlation between the number of mental health facilities and incarceration rates, lending some support to this hypothesis. However, upon performing our linear regression analyses, we found that this relationship was only significant when not accounting for income and unemployment rate, as these social-economic factors dwarf the effects of number of mental health institutions. These cases suggest that access to mental health care is just one piece of a much larger puzzle, and the observed effects of mental health facilities may in part reflect underlying socioeconomic dynamics. However, more analysis needs to be done to parse apart the relationships we found. 

While our findings indicate that expanding mental health treatment infrastructure may contribute to reducing incarceration, the relationship is not definitive. Limitations in our dataset, particularly its scope and size, prevented us from fully isolating the impact of mental health care from other factors. Future research should incorporate a broader dataset, potentially expanding beyond California to include national data, in order to find more powerful insights. Additionally, measuring access solely by facility count may overlook critical aspects such as patient capacity, funding, quality of care, and actual accessibility. For instance, hospitals, outpatient clinics, and telehealth services also function as mental health resources, but were beyond the reach of our dataset.

Ultimately, our analysis highlights that economic and social factors often overshadow the influence of mental health facility count alone. Future studies should seek to define "access" more comprehensively, accounting for patient capacity, funding levels, and barriers to care such as cost, stigma, and insurance coverage. By doing so, we can better assess whether increased access to mental health services truly diverts vulnerable populations away from incarceration and toward the support systems they need. This remains a critical area of study for informing policies aimed at breaking the cycle of incarceration and providing more effective, humane alternatives. With deeper research and a fuller understanding of this relationship, we can hope for a future where those in need are housed in mental health institutions, not jails, receiving care instead of punishment, and ending this unseen housing crisis once and for all.
